```python
import os
from kaggle.api.kaggle_api_extended import KaggleApi
```

```python
import kaggle_variables as kaggle_variables
```

```python
os.environ['KAGGLE_USERNAME'] = kaggle_variables.USERNAME
os.environ['KAGGLE_KEY'] = kaggle_variables.API_TOKEN

api = KaggleApi()
api.authenticate()
```

```python
def download_dataset(dataset):
    api.dataset_download_files(dataset, path=f'raw/{dataset.split("/")[1]}/', unzip=True)
```

```python
download_dataset(kaggle_variables.DATASET1)
download_dataset(kaggle_variables.DATASET2)
download_dataset(kaggle_variables.DATASET3)
download_dataset(kaggle_variables.DATASET4)
```
